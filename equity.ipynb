```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import tree
from sklearn.metrics import confusion_matrix as cm
from sklearn import ensemble
from sklearn.ensemble import GradientBoostingClassifier
from sklearn import linear_model
```

```python
df = pd.read_csv('equitydata.csv')
```

```python
df
```

```python
#selecting all rows, few columns (through feature selection which is already done)
df = df.loc[:, ["Class", "Dividend per Share", "10Y Net Income Growth (per Share)", "10Y Shareholders Equity Growth (per Share)"]]
```

```python
df
```

```python
df = df.dropna()
```

```python
df.shape
```

```python
x = df.loc[:,["Dividend per Share", "10Y Net Income Growth (per Share)", "10Y Shareholders Equity Growth (per Share)"]]
```

```python
y = df.loc[:, ["Class"]]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y)
```

DecisionTreeClassifier

```python
model = tree.DecisionTreeClassifier(max_depth = 4)
```

```python
model.fit(x_train, y_train)
```

```python
pred = model.predict(x_train)
```

```python
cm1 = cm(y_train, pred)
```

```python
cm1
```

```python
accuracy = (cm1[0,0] + cm1[1,1])/(sum(sum(cm1)))
print(accuracy)
```

```python
pred2 = model.predict(x_test)
cm2 = cm(y_test, pred2)
cm2
```

```python
accuracy = (cm2[0,0] + cm2[1,1])/(sum(sum(cm2)))
print(accuracy)
```

-------------------------------------splitting the train test----------------------------------------------

```python
x_train1, x_test1, y_train1, y_test1 = train_test_split(x_train,y_train)
```

```python
for i in range(2,20):
    model3 = tree.DecisionTreeClassifier(max_depth = i)
    model3.fit(x_train1, y_train1)
    pred3 = model3.predict(x_train1)
    cm2 = cm(y_train1, pred3)
    accuracy = (cm2[0,0] + cm2[1,1])/(sum(sum(cm2)))
    print(i)
    print('======================')
    print(accuracy)
```

```python
for i in range(10,300):
    model3 = tree.DecisionTreeClassifier(min_samples_split=i)
    model3.fit(x_train1, y_train1)
    pred3 = model3.predict(x_train1)
    cm2 = cm(y_train1, pred3)
    accuracy = (cm2[0,0] + cm2[1,1])/(sum(sum(cm2)))
    print(i)
    print('======================')
    print(accuracy)
    print()
```

RandomForestClassifier

```python
model4 = ensemble.RandomForestClassifier(max_depth = 6)
```

```python
model4.fit(x_train, y_train)
```

```python
pred4 = model4.predict(x_test)
cm3 = cm(y_test, pred4)
cm3
```

```python
accuracy = (cm3[0,0] + cm3[1,1])/(sum(sum(cm3)))
print(accuracy)
```

GBC

```python
model5 = ensemble.GradientBoostingClassifier(max_depth = 2, min_samples_split = 350)
```

```python
model5.fit(x_train, y_train)
```

```python
pred5 = model5.predict(x_test)
cm4 = cm(y_test, pred5)
cm4
```

```python
accuracy = (cm4[0,0] + cm4[1,1])/(sum(sum(cm4)))
print(accuracy)
```

Regression

```python
model6.fit(x_train, y_train)
```

```python
pred6 = model6.predict(x_test)
cm5 = cm(y_test, pred6)
cm5
```
